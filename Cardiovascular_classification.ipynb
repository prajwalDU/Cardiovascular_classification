# **Problem Description**

**The dataset is from an ongoing cardiovascular study on residents of the town of Framingham, Massachusetts. The classification goal is to predict whether the patient has a 10-year risk of future coronary heart disease (CHD). The dataset provides the patientsâ€™ information. It includes over 4,000 records and 17 attributes. Each attribute is a potential risk factor. There are both demographic, behavioral, and medical risk factors.**

# **Data Description**

**Demographic:**         
â€¢ Sex: male or female("M" or "F")         
â€¢ Age: Age of the patient;(Continuous - Although the recorded ages have been truncated to whole numbers, the concept of age is continuous)     

**Behavioral**    
â€¢ is_smoking: whether or not the patient is a current smoker ("YES" or "NO")  
â€¢ Cigs Per Day: the number of cigarettes that the person smoked on average in one day.(can be considered continuous as one can have any number of cigarettes, even half a cigarette.)     

**Medical( history)**          
â€¢ BP Meds: whether or not the patient was on blood pressure medication (Nominal)         
â€¢ Prevalent Stroke: whether or not the patient had previously had a stroke (Nominal)         
â€¢ Prevalent Hyp: whether or not the patient was hypertensive (Nominal)     
â€¢ Diabetes: whether or not the patient had diabetes (Nominal)

**Medical(current)**      
â€¢ Tot Chol: total cholesterol level (Continuous)       
â€¢ Sys BP: systolic blood pressure (Continuous)         
â€¢ Dia BP: diastolic blood pressure (Continuous)      
â€¢ BMI: Body Mass Index (Continuous)        
â€¢ Heart Rate: heart rate (Continuous - In medical research, variables such as heart rate though in fact discrete, yet are considered continuous because of large number of possible values.)          
â€¢ Glucose: glucose level (Continuous)               

**Predict variable (desired target)**         
â€¢ 10-year risk of coronary heart disease CHD(binary: â€œ1â€, means â€œYesâ€, â€œ0â€ means â€œNoâ€) - DV

Consider age column remaining all columns are contains outliers so we want treat it with median using interquartile range method

```python
# Copying our dataset so that we compare 
data = df.copy()
# Finding lower and upper limit
for col in numerical_features:
  q1, q3, median = data[col].quantile([0.25,0.75,0.5])
  lower_limit = q1 - 1.5*(q3-q1)
  upper_limit = q3 + 1.5*(q3-q1)

  # Replacing outliers with median value
  data[col] = np.where(data[col] > upper_limit, median, np.where(
                      data[col] < lower_limit, median, data[col]))
```

```python
# Plotting graphs before treating outliers
for col in numerical_features:
  fig, ax = plt.subplots(1,4, constrained_layout=True)
  fig.set_size_inches(15,3)
  sns.distplot(df[col], ax=ax[0]).set(title='Before')
  sns.boxplot(df[col], ax=ax[1]).set(title='Before')
  # Plotting graphs after treating outliers
  sns.distplot(data[col], ax=ax[2]).set(title='After')
  sns.boxplot(data[col], ax=ax[3]).set(title='After')
  plt.suptitle(f'{col.title()}', weight='bold')
  plt.show()
```

## **Checking Multicolinerity**

```python
# Write a function for varaiance inflation factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Defining a VIF function
def calc_vif(X):
  # Calculating VIF
  vif = pd.DataFrame()
  vif['variables'] = X.columns
  vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
  return(vif.sort_values(by='VIF', ascending=False).reset_index(drop=True)) 
```

# **Model Building**

```python
# Importing packages
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, roc_curve
from sklearn.metrics import f1_score, make_scorer, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBRFClassifier
from imblearn.combine import SMOTETomek
```

```python
# Scaling variables
scaler = MinMaxScaler()
scoring = make_scorer(f1_score, pos_label=1)
features = [i for i in data.columns if i not in ['TenYearCHD']]
```

```python
# Spliting dataset into train and test dataset
X = scaler.fit_transform(data[independent_variables])
y = data['TenYearCHD']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
# Handling class imbalance by Oversampling followed by removing tomek link
X_smote, y_smote = SMOTETomek(random_state=42).fit_resample(X_train, y_train)

# Check class imbalance both before and after class imbalance
for col,label in [[y_train,'Before'],[y_smote,'After']]:
  print(label+'Handling Class Imbalance:')
  print(col.value_counts(),'\n')
```

```python
# Define a function to train the input model and print evalution matrix
def analysis_model(model, X_train, X_test, y_train, y_test):
  '''
  Fitting model to test multiple algorithms to find which algorithms gives best result,
  plotting Classification report,AUC_ROC curve,Confusion matrix
  finally we plot the feature importance
  '''
  # Fitting the model
  model.fit(X_train,y_train)

  # Feature importance
  try:
    try:
      importance = model.feature_importance_
      feature = features
    except:
      importance = np.abs(model.coef_[0])
      feature = independent_variables
    indices = np.argsort(importance)
    indices = indices[::-1]
  except:
    pass

  # Plot evaluation metrices for train and test dataset
  for x, act, label in ((X_train,y_train,'Train_set'),(X_test,y_test,'Test_set')):

    # Getting required metrics
    pred = model.predict(x)
    pred_proba = model.predict_proba(x)[:,1]
    report = pd.DataFrame(classification_report(y_pred=pred, y_true=act, output_dict=True))
    fpr, tpr, threholds = roc_curve(act,pred_proba)

    # Classification report
    plt.figure(figsize=(18,3))
    plt.subplot(1,3,1)
    sns.heatmap(report.iloc[:-1, :-1].T, annot=True, cmap='coolwarm')
    plt.title(f'{label} Confusion Matrix')
    plt.xlabel('Predicted labels')
    plt.ylabel('Actual labels')

    # AUC_ROC curve
    plt.subplot(1,3,2)
    plt.plot([0,1],[0,1], 'k--')
    plt.plot(fpr, tpr, label=f'AUC = {np.round(np.trapz(tpr,fpr),3)}')
    plt.legend(loc=4)
    plt.title(f'{label} AUC_ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    # Confusion matrix
    plt.subplot(1,3,3)
    sns.heatmap(confusion_matrix(y_true=act, y_pred=pred), annot=True, cmap='coolwarm')
    plt.title(f'{label} Confusion Matrix')
    plt.xlabel('Predicted labels')
    plt.ylabel('Actual labels')
    plt.tight_layout()

  # plot Feature importance
  try:
    plt.figure(figsize=(18,3))
    plt.bar(range(len(indices)), importance[indices])
    plt.xticks(range(len(indices)), [feature[i] for i in indices])
    plt.title('Feature Importance')
    plt.tight_layout()
  except:
    pass
  plt.show()

  return model
```

### **Logistic Regression**

```python
# Fit logistic regression
lr = LogisticRegression(fit_intercept=True, max_iter=10000)
analysis_model(lr, X_smote, X_test, y_smote, y_test)
```

### **Naive Bayes Classifier**

```python
# Fit naive bayes classifier
nbc = GaussianNB()
analysis_model(nbc, X_smote, X_test, y_smote, y_test)
```

### **Support vector Classifier**



```python
# SVC algorithm
clf = SVC(random_state=0, probability=True)

# Hyperparameter grid
grid = {'kernel' : ["linear","rbf","poly","sigmoid"],
        'C' : [0.1, 1, 10, 100],
        'max_iter' : [1000]}

# GridSearchCV to find best parameters
svc = GridSearchCV(clf, param_grid=grid, scoring=scoring, cv=5)
svc.fit(X_smote, y_smote)

# Analysising model with best parameters
analysis_model(svc.best_estimator_, X_smote, X_test, y_smote, y_test) 
```

### **Random Forest Classifier**

```python
# Split dataset into train and test by redefining X
X = scaler.fit_transform(data[features])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Handling call imbalance by oversampling followed by tomek link
X_smote, y_smote = SMOTETomek(random_state=45).fit_resample(X_train,y_train)
```

### **XGBoost Classifier**


```python
# Classifier
clf = XGBRFClassifier(silent=True, random_state=3)

# Hyparparameter grid
grid = {'n_estimators' : [150],
        'max_depth' : [8,10],
        'min_samples_split' : [50,100],
        'min_samples_lead' : [30,50],
        'eta' : [0.05,0.08,0.1]}

# GridSearchCV to find best parameters
xgb = GridSearchCV(clf, param_grid=grid, scoring=scoring, cv=5)
xgb.fit(X_smote,y_smote)

# Analysing model with best set of parameters
analysis_model(xgb.best_estimator_, X_smote, X_test, y_smote, y_test)
```

### **KNN Classifer**

# **Conclusion**

If we want to completely avoid any situations where the patient has heart disease, a high recall is desired. Whereas if we want to avoid treating a patient with no heart diseases a high precision is desired.          
Assuming that in our case the patients who were incorrectly classified as suffering from heart disease are equally important since they could be indicative of some other ailment, so we want a balance between precision and recall and a high f1 score is desired.                      
Since we have added synthetic datapoints to handle the huge class imbalance in training set, the data distribution in train and test are different so the high performance of models in the train set is due to the train-test data distribution mismatch and not due to overfitting.                 
Best performance of Models on test data based on evaluation metrics for class 1:
Recall - SVC                    
Precision - Naive Bayes Classifier                  
F1 Score - Logistic Regression, XGBoost
