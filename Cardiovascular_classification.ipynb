# **The dataset is from an ongoing cardiovascular study on residents of the town of Framingham, Massachusetts. The classification goal is to predict whether the patient has a 10-year risk of future coronary heart disease (CHD). The dataset provides the patientsâ€™ information. It includes over 4,000 records and 15 attributes. Each attribute is a potential risk factor. There are both demographic, behavioral, and medical risk factors.**

Cardiovascular_disease_3.png

# **Data Description**
**Demographic:**         
â€¢ Sex: male or female("M" or "F")         
â€¢ Age: Age of the patient;(Continuous - Although the recorded ages have been truncated to whole numbers, the concept of age is continuous)     

**Behavioral**    
â€¢ is_smoking: whether or not the patient is a current smoker ("YES" or "NO")  
â€¢ Cigs Per Day: the number of cigarettes that the person smoked on average in one day.(can be considered continuous as one can have any number of cigarettes, even half a cigarette.)     

**Medical( history)**          
â€¢ BP Meds: whether or not the patient was on blood pressure medication (Nominal)         
â€¢ Prevalent Stroke: whether or not the patient had previously had a stroke (Nominal)         
â€¢ Prevalent Hyp: whether or not the patient was hypertensive (Nominal)     
â€¢ Diabetes: whether or not the patient had diabetes (Nominal)

**Medical(current)**      
â€¢ Tot Chol: total cholesterol level (Continuous)       
â€¢ Sys BP: systolic blood pressure (Continuous)         
â€¢ Dia BP: diastolic blood pressure (Continuous)      
â€¢ BMI: Body Mass Index (Continuous)        
â€¢ Heart Rate: heart rate (Continuous - In medical research, variables such as heart rate though in fact discrete, yet are considered continuous because of large number of possible values.)          
â€¢ Glucose: glucose level (Continuous)               

**Predict variable (desired target)**         
â€¢ 10-year risk of coronary heart disease CHD(binary: â€œ1â€, means â€œYesâ€, â€œ0â€ means â€œNoâ€) - DV

# **Handling Outliers**

```python
# Visualizing the distributions
plt.figure(figsize=(18,12))
for n, column in enumerate(numerical_features):
  plt.subplot(5, 4, n+1)
  sns.distplot(df[column])
  plt.title(f'{column.title()}', fontweight='bold')
  plt.tight_layout()
```

```python
# Visiualizing boxplot is easyway to find outliers in our dataset
plt.figure(figsize=(18,12))
for n, column in enumerate(numerical_features):
  plt.subplot(5,4, n+1)
  sns.boxplot(df[column])
  plt.title(f'{column.title()}', weight='bold')
  plt.tight_layout()
```

```python
# Removing outliers by IQR method
data = df.copy()
for col in numerical_features:
  q1, q3, median = data[col].quantile([0.25,0.75,0.5])
  lower_limit = q1 - 1.5*(q3-q1)
  upper_limit = q3 + 1.5*(q3-q1)

  # Replacing outliers with median value
  data[col] = np.where(data[col] > upper_limit, median, np.where(
                      data[col] < lower_limit, median, data[col]))
```

```python
# Plotting graphs before treating outliers
for col in numerical_features:
  fig, ax = plt.subplots(1,4, constrained_layout=True)
  fig.set_size_inches(15,3)
  sns.distplot(df[col], ax=ax[0]).set(title='Before')
  sns.boxplot(df[col], ax=ax[1]).set(title='Before')
  # Plotting graphs after treating outliers
  sns.distplot(data[col], ax=ax[2]).set(title='After')
  sns.boxplot(data[col], ax=ax[3]).set(title='After')
  plt.suptitle(f'{col.title()}', weight='bold')
  plt.show()
```

# **Checking Multicolinerity**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Defining a VIF function
def cal_vif(X):
  # Calculating VIF
  vif = pd.DataFrame()
  vif['variables'] = X.columns
  vif['VIf'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
  return(vif.sort_values(by='VIF', ascending=False).reset_index(drop=True)) 
```
